```python
def generate_ner(data):
    sentence = data['sentence']
    ner = data['ner']
    ner_sorted = {k: v for k, v in sorted(ner.items(), key=lambda item: item[1])}
    
    
    list_ner = []
    list_sentence = []
    max_len = len(ner_sorted)
    for i,k in enumerate(ner_sorted):
        v = ner_sorted[k]
        if i == 0:
            end = v[0]
            sent = sentence[0:end].strip().split(' ')
            for word in sent:
                if word != ' ':
                    list_ner.append('O')
                    list_sentence.append(word)
        
        #ner
        sent = sentence[v[0]:v[1]].strip().split(' ')
        for index,word in enumerate(sent):
            if index == 0:
                list_ner.append(f'B_{k}')
            else:
                list_ner.append(f'I_{k}')
            list_sentence.append(word)
        
        if i == max_len - 1:
            begin = v[1]
            sent = sentence[begin:].strip().split(' ')
            for word in sent:
                if word != ' ':
                    list_ner.append('O')
                    list_sentence.append(word)
    
    return list_ner, list_sentence
```
