```python
prompt = """
Dựa vào ngữ cảnh bạn đã cung cấp, dưới đây là 5 câu hỏi ngắn gọn liên quan:

Tỷ lệ xử lý đúng hạn của VTPOST trong tháng 4 năm 2021 là bao nhiêu?
Mục tiêu KPI ban đầu cho tỷ lệ xử lý đúng hạn của VTPOST trong tháng 4 năm 2021 là gì?
Tỷ lệ xử lý đúng hạn của VTPOST trong tháng 4 năm 2021 so với mục tiêu KPI ban đầu chênh lệch như thế nào?
Tỷ lệ xử lý đúng hạn của VTPOST trong tháng 4 năm 2021 đã tăng bao nhiêu phần trăm so với KPI của tháng 3 năm 2021?
Tỷ lệ xử lý đúng hạn của VTPOST trong tháng 4 năm 2021 đã đạt mục tiêu KPI ban đầu hay không?
Lưu ý: Các câu hỏi này dựa trên ngữ cảnh cung cấp và không liên quan trực tiếp đến cấu trúc list của Python.

"""
```

```python
def get_question(text):
  question_check = prompt.split("\n")
  list_res = []
  for question in question_check:
    if '?' in question:
      if question[0:2] in ["1.","2.","3.","4.","5."]:
        list_res.append(question[2:])
      else:
        list_res.append(question)

  return list_res
```
