```python
chunks = splitter.split_text(context)
```

```python
def infer(examples):
    model_mrc.eval()

    template = f'nội dung:{examples["context"]}\ncâu hỏi:{examples["question"]}'
    inputs = tokenizer_mrc(template, max_length=512, truncation=True, padding='max_length', return_tensors='pt')
    output = model_mrc.generate(**inputs, num_beams=8, do_sample=True, max_length=128)

    decoded_output = tokenizer_mrc.batch_decode(output, skip_special_tokens=True)[0]

    return decoded_output
```

```python
from sentence_transformers import SentenceTransformer, util

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
model_rerank = SentenceTransformer('keepitreal/vietnamese-sbert').to(device)
```

```python
def query_sim(sentence,answers):
    embedding = model_rerank.encode(answers, convert_to_tensor=True)
    query_embedding = model_rerank.encode(sentence, convert_to_tensor=True).to(device)
    cos_scores = util.cos_sim(query_embedding, embedding)[0]
    top_results = torch.topk(cos_scores, k=1)

    return answers[int(top_results[1])]
```
