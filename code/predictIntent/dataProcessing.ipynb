```python
l2i = {ele.split('.')[0] : i for i,ele in enumerate(os.listdir(link_data)) if ele.split('.')[0] != 'childQuestion'}
i2l = {i : ele.split('.')[0] for i,ele in enumerate(os.listdir(link_data)) if ele.split('.')[0] != 'childQuestion'}
```

```python
l2i
```

```python
import json
import pickle
from tqdm import tqdm

texts = []
labels = []

for name in tqdm(os.listdir(link_data)):
    prefix_name = name.split('.')[0]
    if prefix_name == "childQuestion":
      continue
    with open(link_data + name,'rb') as f:
      data = pickle.load(f)
    for ele in data:
      texts.append(ele)
      labels.append(l2i[prefix_name])
```
